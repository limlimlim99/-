```python
import urllib.request as req
```

```python
data=req.urlopen("http://daum.net").read().decode("utf-8")
```

```python
data
```

```python
import re
p=re.compile("^https://.js$")
p.findall(data)
for i in p.findall(data):
    print(i)
```

```python
p=re.compile("https[:./\w]*[.]js")

for i in p.findall(data):
    print(i)
```

```python
p = re.compile("^https://\S+.js$")
li = re.findall(p, data)
for i in li:
    print(i) 
```
