```python
# 구간 나누기, rank열을 추가
# ~20 : A
# 20~30 : B
# 30~40 : c
# 40이상 : D
mpg.hwy.describe()
```

```python
def grad(x):
    if x < 20:
        return 'A'
    elif x < 30:
        return 'B'
    elif x < 40:
        return 'C'
    elif x >= 40:
        return 'D'

mpg['rank'] = mpg['hwy'].map(grad)
mpg
```
