```python
with open('speech_moon.txt', 'r', encoding = 'utf-8') as f:
    moon = f.read()
```

```python
moon
```

```python
import re
p=re.compile("[^가-힣]")
moon = p.sub(" ",moon)
moon
```

```python
import konlpy
hannanum=konlpy.tag.Hannanum()
```

```python
hannanum.nouns("대한민국 국민입니다")
```

```python
nouns=hannanum.nouns(moon)
nouns
```

```python
import pandas as pd
df=pd.DataFrame(nouns)
df[:2]
```

```python
import pandas as pd
df_word=pd.DataFrame({'word':nouns})
df_word
```

```python
df_word['count']=df_word['word'].str.len()
df_word
```

```python
df_word=df_word.query('count>=2')
df_word.sort_values('count')
```

```python
df_word=df_word.groupby('word',as_index=False).agg(n=('word','count')).sort_values('n',ascending=False)
df_word
```

```python
top20=df_word.head(20)
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams.update({'font.family':'Malgun Gothic'})

sns.barplot(data=top20,y='word',x='n')
```
