- 파이썬과 넘파이 계산 속도 비교

```python
list1=[1,2,3,4]
list2=[1,2,3,4]
```

```python
list3=[]
for i in range(4):
    list3.append(list1[i]+list2[i])
list3
```

```python
array1=np.array([1,2,3,4])
array2=np.array([1,2,3,4])
```

```python
array3=array1+array2
array3
```

```python
def add_list(n):
    li1=list(range(1,n+1))
    li2=list(range(1,n+1))
    
    li3=[]
    for i in range(n):
        li3.append(li1[i]+li2[i])
        
    return li3

add_list(10)
```

```python
def add_array():
    array1=np.array(range(1000))
    array2=np.array(range(1000))
    array3=array1+array2
        
    return array3

add_array()
```

```python
np.array(range(1,10))
```

```python
def add_list():
    li1=list(range(10000))
    li2=list(range(10000))
    
    li3=[]
    for i in range(10000):
        li3.append(li1[i]+li2[i])
        
    return li3

add_list()
```

```python
import timeit
```

```python
timeit.timeit(add_list,number=10)
```

```python
print(timeit.timeit(add_list,number=1000))
print(timeit.timeit(add_array,number=1000))
```
